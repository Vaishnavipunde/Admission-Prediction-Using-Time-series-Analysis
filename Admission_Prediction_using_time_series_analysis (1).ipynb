# Outlier removing techniques

Trimming:

Trimming involves removing a certain percentage of the extreme values from both ends of the dataset.
Trimming removes the most extreme values, reducing the impact of outliers on statistical measures. However, it also reduces the size of the dataset, potentially affecting the representativeness of the data.

Replacement:

Replacement involves replacing outlier values with more reasonable estimates.
Outliers can be replaced with other values, such as the mean, median.

Winsorization:

Winsorization limits extreme values by replacing them with less extreme values within a certain range.
Instead of removing outliers, winsorization replaces extreme values with values from the edge of a specified range.
'''

# Data stationarity
Augmented Dickey-Fuller (ADF) test, a statistical test used to determine whether a given time series data is stationary or not. Stationarity implies that the statistical properties of the time series, such as mean, variance, and autocorrelation, do not change over time. 
The null hypothesis (H0) of the ADF test is that the time series has a unit root, indicating it is non-stationary. If the p-value obtained from the test is less than a chosen significance level (such as 0.05), we reject the null hypothesis, indicating that the time series is likely stationary. 


The ADF test returns a few key components:

ADF Statistic: This is the test statistic obtained from the ADF test. It is used in conjunction with critical values to determine the stationarity of the series. More negative values indicate stronger evidence against the presence of a unit root.

p-value: The probability of observing the ADF statistic if the null hypothesis is true. A lower p-value suggests stronger evidence against the null hypothesis.

Critical Values: These are the critical values at different confidence levels (1%, 5%, and 10%). Comparing the ADF statistic with these critical values helps in determining stationarity.

# Bhojpur

```python
#generate a line plot for the 'Bhojpur' column from the DataFrame data using Matplotlib.
# Plotting the 'Bhojpur' column
data['Bhojpur'].plot()
# Setting the title of the plot
plt.title('Bhojpur') 
# Setting the label for the x-axis
# Setting the label for the y-axis
plt.xlabel('Year of admission') 
plt.ylabel('Number of students')
# Displaying the plot
plt.show() 
```

```python
#for predicting data is stationary or not we have to do ADF test.
#In ADF test we calculate p-value.p-value is depend on probability distribution.
#The P value is defined as the probability under the assumption of null hypothesis.
#A null hypothesis is a type of statistical hypothesis 
#A null hypothesis proposes that no statistical significance exists in a set of given observations.
#if P-value is less than 0.05 then the data is stationary.

#import adfuller from stats models
from statsmodels.tsa.stattools import adfuller

#performance of augumented dicky fuller test
print("Result of Dicky fuller test")
result=adfuller(data['Bhojpur'],autolag='AIC')
print("ADF statistics: %f"%result[0])
print("p-value:   %f"%result[1])
print("crtical values")
for key,value in result[4].items():
    print(key,value)
if result[1] <= 0.05:
    print('reject the null hypothesis')
    print("as the p-value is less than 0.05 the time series is stationary")
else:
    print('Not Stationary')
    print('fail to reject the null hypothesis')

#here we can see that p-value is 0.006839 which is less than 0.05
#so the data is stationary.
```

```python
#for predicting data is stationary or not we have to do ADF test.
#In ADF test we calculate p-value.p-value is depend on probability distribution.
#The P value is defined as the probability under the assumption of null hypothesis.
#A null hypothesis is a type of statistical hypothesis 
#A null hypothesis proposes that no statistical significance exists in a set of given observations.
#if P-value is less than 0.05 then the data is stationary.
#import adfuller from stats models
from statsmodels.tsa.stattools import adfuller

#performance of augumented dicky fuller test
print("Result of Dicky fuller test")
result=adfuller(data['Hirapur'],autolag='AIC')
print("ADF statistics: %f"%result[0])
print("p-value:   %f"%result[1])
print("crtical values")
for key,value in result[4].items():
    print(key,value)
if result[1] <= 0.05:
    print('reject the null hypothesis')
    print("as the p-value is less than 0.05 the time series is stationary")
else:
    print('Not Stationary')
    print('fail to reject the null hypothesis')
#here we can see that p-value is 0.198156 which is greater than 0.05
#so the data is non-stationary.        
```

```python
#Perform shifting of data 
data['Hirapur_first_diff']=data['Hirapur']-data['Hirapur'].shift(1)
```

```python
#Again check the p-value after shifting
#import adfuller from stats models
from statsmodels.tsa.stattools import adfuller

#performance of augumented dicky fuller test
print("Result of Dicky fuller test")
result=adfuller(data['Hirapur_first_diff'].dropna(),autolag='AIC')
print("ADF statistics: %f"%result[0])
print("p-value:   %f"%result[1])
print("crtical values")
for key,value in result[4].items():
    print(key,value)
if result[1] <= 0.05:
    print('reject the null hypothesis')
    print("as the p-value is less than 0.05 the time series is stationary")
else:
    print('Not Stationary')
    print('fail to reject the null hypothesis')
#here we can see that p-value is 0.000000 which is smaller than 0.05
#so the data is stationary.  
```

# Model Selection
ARIMA (Autoregressive Integrated Moving Average) and SARIMA (Seasonal Autoregressive Integrated Moving Average) models are popular time series forecasting techniques used to make predictions based on historical data.

Here's a brief explanation of these models:

ARIMA Model:

ARIMA models are based on the combination of three components:
Autoregression (AR): The model uses the dependent relationship between an observation and a number of lagged observations.
Differencing (I): A method of differencing to make the time series stationary (i.e., removing trend and seasonality).
Moving Average (MA): The model accounts for the relationship between an observation and a residual error from a moving average model applied to lagged observations.
Parameters of an ARIMA model include p, d, and q, which represent the order of the AR, differencing, and MA components respectively.


SARIMA Model:

SARIMA extends ARIMA to handle seasonal variations in time series data.
It includes additional parameters for seasonal components, denoted as P, D, Q, and m, representing the seasonal order and the period of seasonality.
